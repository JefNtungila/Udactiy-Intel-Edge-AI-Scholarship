```
from imageai.Detection import VideoObjectDetection
import os

def convert_avi_to_mp4(avi_file_path, output_name):
    os.popen("ffmpeg -i '{input}' -ac 2 -b:v 2000k \
              -c:a aac -c:v libx264 -b:a 160k \
              -vprofile high -bf 0 -strict experimental \
              -f mp4 '{output}.mp4'"\
             .format(input = avi_file_path, 
                     output = output_name))
    return True

def detect(input_filename):
    execution_path = os.getcwd()
    detector = VideoObjectDetection()
    detector.setModelTypeAsYOLOv3()
    detector.setModelPath(os.path.join(execution_path, "yolo.h5"))
    detector.loadModel()
    output_filename = input_filename.split('.')[0]+'_detected'
    video_path = detector.detectObjectsFromVideo(
        input_file_path=os.path.join(execution_path, input_filename),
        output_file_path=os.path.join(execution_path, output_filename), 
        frames_per_second=29, 
        log_progress=True)
    convert_avi_to_mp4(output_filename+'.avi', output_filename)
```
